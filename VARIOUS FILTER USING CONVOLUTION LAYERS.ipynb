```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imshow , imread
from skimage.color import rgb2yuv,rgb2hsv,rgb2gray,yuv2rgb,hsv2rgb
from scipy.signal import convolve2d
```

```python
goku = imread(r"C:\Users\mitesh salunkhe\OneDrive\Pictures\Screenshots\Screenshot (45).png")[:, :, :3]
plt.figure(num=None,figsize=[10,10],dpi=80)
imshow(goku)
```

```python
#filters
sharpen = np.array([[0,-1,0],
                   [-1,4,-1],
                   [0,-1,0]])
blur =np.array([[0.11,0.11,0.11],
               [0.11,0.11,0.11],
               [0.11,0.11,0.11]])

laplacian = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

vertical = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])

horizontal = np.array([[-1,2,-1],
                       [0,0,0],
                       [1,2,1]])

gaussian = (1/16.0) * np.array([[1,2,1],
                               [2,4,2],
                               [1,2,1]])
```

```python
fig,ax = plt.subplots(1,6, figsize = (10,5))

ax[0].imshow(sharpen,cmap = "gray")
ax[0].set_title(f"sharpen" , fontsize = 20)

ax[1].imshow(laplacian,cmap = "gray")
ax[1].set_title(f"laplacian" , fontsize = 20)

ax[2].imshow(vertical,cmap = "gray")
ax[2].set_title(f"vertical" , fontsize = 20)

ax[3].imshow(horizontal,cmap = "gray")
ax[3].set_title(f"horizontal" , fontsize = 20)

ax[4].imshow(gaussian,cmap = "gray")
ax[4].set_title(f"gausian" , fontsize = 20)

ax[5].imshow(blur,cmap = "gray")
ax[5].set_title(f"blur" , fontsize = 20)

plt.tight_layout()
plt.show()
```
